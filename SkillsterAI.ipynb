```python
import mediapipe as mp
import cv2
from google.colab.patches import cv2_imshow
from gtts import gTTS
import IPython.display as ipd

# Initialize MediaPipe Tennis Pose model
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

# Function to draw landmarks with annotations
def draw_landmarks(image, landmarks):
    annotated_image = image.copy()
    for idx, landmark in enumerate(landmarks.landmark):
        # Convert landmark index to body part name
        body_part = mp_pose.PoseLandmark(idx).name
        # Draw landmark on image
        h, w, _ = image.shape
        cx, cy = int(landmark.x * w), int(landmark.y * h)
        cv2.circle(annotated_image, (cx, cy), 5, (255, 0, 0), -1)
        # Add annotation text
        cv2.putText(annotated_image, body_part, (cx + 10, cy), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1, cv2.LINE_AA)
    return annotated_image

# Load the professional tennis serve image
professional_image_path = "proserve.jpg"
professional_image = cv2.imread(professional_image_path)
professional_image_rgb = cv2.cvtColor(professional_image, cv2.COLOR_BGR2RGB)

# Initialize MediaPipe Pose object for professional pose
with mp_pose.Pose(static_image_mode=True, min_detection_confidence=0.5) as pose:
    # Process the professional tennis serve image
    professional_results = pose.process(professional_image_rgb)
    if professional_results.pose_landmarks:
        professional_landmarks = professional_results.pose_landmarks

# Load the user input image
user_input_image_path = "serve.jpg"
user_input_image = cv2.imread(user_input_image_path)
user_input_image_rgb = cv2.cvtColor(user_input_image, cv2.COLOR_BGR2RGB)


# Process the user input image with MediaPipe
with mp_pose.Pose(static_image_mode=True, min_detection_confidence=0.5) as pose:
    results = pose.process(user_input_image_rgb)
    if results.pose_landmarks:
        user_pose_landmarks = results.pose_landmarks
        # Draw user pose landmarks on the image
        user_annotated_image = draw_landmarks(user_input_image_rgb, user_pose_landmarks)
        # Display the image
        cv2_imshow(user_annotated_image)


# Compare the user's pose with the professional's pose
if user_pose_landmarks and professional_landmarks:
    diff_threshold = 0.05  # Adjust threshold as needed
    similar_body_parts = []
    for idx, user_landmark in enumerate(user_pose_landmarks.landmark):
        professional_landmark = professional_landmarks.landmark[idx]
        # Calculate Euclidean distance between user and professional landmarks
        dist = ((user_landmark.x - professional_landmark.x) ** 2 + (user_landmark.y - professional_landmark.y) ** 2) ** 0.5
        # Compare distance to threshold
        if dist < diff_threshold:
            similar_body_parts.append(mp_pose.PoseLandmark(idx).name)

    if similar_body_parts:
        # Check hand positions in both images
        user_left_hand_y = user_pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y
        user_right_hand_y = user_pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].y
        professional_left_hand_y = professional_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y
        professional_right_hand_y = professional_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].y
    # Check if professional's hand is more up than the user's hand
if professional_left_hand_y < user_left_hand_y or professional_right_hand_y < user_right_hand_y:
    comment = "Raise your left hand higher for achieving the accurate posture for this pose and try again."
elif professional_right_hand_y < user_right_hand_y:
    comment = "Raise your left hand higher for achieving the accurate posture for this pose"
else:
    comment = "Your pose is similar to the professional's pose in the following body parts: " + ', '.join(similar_body_parts)

print("Comparison Result:", comment)

# Convert text comment into speech
tts = gTTS(comment)
tts.save("comment.mp3")

# Play the audio comment
ipd.Audio("comment.mp3", autoplay=True)



```
